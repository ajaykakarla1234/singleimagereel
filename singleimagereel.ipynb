```python
# Step 1: Collect User Input
video_idea = "One highly motivational quote for a young man,  one image without any text in it for a single visual section and quote being present in text section"
video_duration = 20  # in seconds

# Step 2: Generate Script
import os
from groq import Groq

client = Groq(
    api_key="YOUR_GROQ_API_KEY",
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"""
Generate a single motivational quote: "{video_idea}". .

### Instructions for Visuals:
- **Ensure that no text appears on the visual** (e.g., no text overlays or subtitles). Visuals should be free from any text.
- The visual should match the mood, theme, and message of the text, with images such as historical depictions, symbolism, and scenery relevant to the text.

Title: "Motivational Quote"

Visual:
A sunrise over a mountain range, with a single hiker standing triumphantly at the peak.
Text:
Theodore Roosevelt once said, 'Believe you can, and you’re halfway there.'
"""
        }
    ],
    model="llama3-8b-8192",
)

script = chat_completion.choices[0].message.content
print(script)

# Step 3: Extract Voiceover Text
import re

# Extract voiceover text from script
voiceover_match = re.search(r"Text:\n(.*?)$", script, re.DOTALL)
voiceover_text = voiceover_match.group(1).strip() if voiceover_match else ""

print(f"Voiceover Text:\n{voiceover_text}\n")

# Step 4: Generate Image with Text
from PIL import Image, ImageDraw, ImageFont

# Create an image
image = Image.new("RGB", (1080, 1080), color=(255, 255, 255))

# Load a font
font_path = "/usr/share/fonts/truetype/humor-sans/Humor-Sans.ttf"
font = ImageFont.truetype(font_path, 40)

# Initialize ImageDraw
draw = ImageDraw.Draw(image)

# Define text and get its size
text = "Theodore Roosevelt once said, 'Believe you can, and you’re halfway there.'"
text_width, text_height = draw.textsize(text, font=font)

# Calculate the position for the text to be centered
x = (image.width - text_width) / 2
y = (image.height - text_height) / 2

# Add text to image
draw.text((x, y), text, font=font, fill="black")

# Save the image
image_path = "/content/motivational_image.png"
image.save(image_path)

print(f"Image saved at {image_path}")

# Step 5: Text-to-Speech for Voiceover
from gtts import gTTS

tts = gTTS(text=voiceover_text, lang='en')
audio_file = "/content/voiceover.mp3"
tts.save(audio_file)

print(f"Voiceover saved at {audio_file}")

# Step 6: Add Background Music and Compile Media
import moviepy.editor as mp

# Load the image and set its duration
image_clip = mp.ImageClip(image_path).set_duration(video_duration).set_fps(24)

# Load the voiceover audio
voiceover_audio = mp.AudioFileClip(audio_file)

# Load background music
background_music_path = "/content/background_music.mp3"
background_music = mp.AudioFileClip(background_music_path).subclip(0, video_duration).volumex(0.5)

# Combine voiceover and background music
final_audio = mp.CompositeAudioClip([voiceover_audio, background_music])

# Set the audio to the image clip
final_video = image_clip.set_audio(final_audio)

# Step 7: Export Video
output_path = "/content/final_video.mp4"
final_video.write_videofile(output_path, codec="libx264", audio_codec="aac", fps=24)

print(f"Video generation complete! Video saved at {output_path}")
```
